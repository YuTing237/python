#錯誤處理
try: <br>
  程式碼

except: <br>
  例外的時候處理動作

else: <br>
  當沒有進入例外(程式正確)的處理動作

finally: <br>
  不管正確例外,都會執行的動作

```python
a=10
b=20
if a<b:
    print(n) #錯誤
```

```python
a=10
b=20
try:
 if a<b:
   print(n) #錯誤
except:
   print("錯誤產生") #發生例外狀況,就會印出"錯誤產生"
```

#指定錯誤類別處理

```python
a=10
b=20
try:
 if a<b:
   print(n) #錯誤
except NameError as e:
   print("錯誤產生, 錯誤原因：",e) #發生例外狀況,就會印出錯誤產生,錯誤原因： name 'n' is not defined
except ZeroDivisionError: #在進行運算時，拿0當成除數時所產生的錯誤
   print("除以0錯誤")
```

#else
沒有錯誤產生會執行的部分

```python
a=10
b=20
try:
    if a>b: #改變比較條件，讓錯誤不發生
        print(n)
except:
    print("錯誤發生") #例外狀況
else:
    print("沒有錯誤發生時會執行的部分") #如果沒有錯誤產生就會執行的
```

```python
try:
  a=10
  b=20
  if a<b:
    print(n)
except NameError:
  print("NameError錯誤產生") #條件是符合的,就會產生這行
except:
  print("其他錯誤產生") #前面產生出來,except和else就不會印出
else:
  print("沒有任何錯誤產生")
finally:
  print("不管程式正確錯誤都會執行的部分") #即使前面已經有執行出來了,這行也一定會印出
```
