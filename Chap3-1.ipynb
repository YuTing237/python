##Tuple (元組)
tuple(元組)可視為不可改變的串列，tuple跟list很像，但是我們不能新增、刪除或更新tuple的元素。tuple的好處有：

1.佔用較少空間 

2.可當作dictionary(字典)的鍵值

3.可當作函式引數

```python
a_tuple=('python', 'js', 'SQL')
print(a_tuple)  #會印出a_tuple的資料
b_list=['go', 'c#', 'vb']  #list可更改 
b_tuple=tuple(b_list)     #tuple不可更改 
print(b_tuple)    #會印出b_tuple的資料
print(b_tuple[0])   #取b_tuple的第0個資料
```

#字典(dictionary - dict)為帶有鍵值(key)的串列(list)

```python
languages = {}
a={0:'python',1:'java',2:'SQL'}
print(a[0]) #取a資料裡的第0個資料
print(a[1]) #取a資料裡的第1個資料
print(a[2]) #取a資料裡的第2個資料

b={"age":47,"gender":"M","name":"suhsien"}
print(b["age"])  #取b資料裡的age設定好的資料
print(b["gender"]) #取b資料裡的gender設定好的資料
print(b["name"]) #取b資料裡的name設定好的資料
```

```python
languages = {}
languages = {'name':'python', 'version':'3.7'} #索引值(鍵值)可以自訂
print(languages['name'])  #取languages裡name要產生的資料
print(languages['version']) #取languages裡version要產生的資料
print(languages)      #取languages裡的所有資料
print(languages.keys())   #keys是方法，用来取值的。
print(languages.values())  #values為值
print(languages.items())  #會將languages裡的資料項目進行分開印出
print('name' in languages)  #印出name是否有在languages裡
```
