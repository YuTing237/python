# 運算式

```python
a=8
b=5
#透過第三變數c,交換a,b的值
c=a
a=b
b=c
print(a) #印出a
print(b) #印出b
```

```python
#不透過第三變數,交換a,b的值
a=8
b=5
a=a+b
b=a-b
a=a-b
print(a) #印出a
print(b) #印出b
```

```python
a=8
b=5
a,b=b,a
#python版本的變數交換
print(a) #印出a
print(b) #印出b
```

# 型態轉換

```python
x="must" #x設為字串
print(10*x) #10*x為印出10個must
print("10"+x) #因為前面是字串後面的x也是字串,所以會產生10must(有加入""的為字串)
print(10,x) #單純印出10與must(會區隔)
```

```python
x=12
print(type(x)) 
y=float(x)+0.5  #強制型別轉換
print(type(y)) 
print(y) #印出y的數字
print(int(y)) #原本y是有小數的,但因為前面加int,所以強制轉換成整數
print(int("12")) #雖然後面是字串,但前面加入了int,所以強制轉換,但原本就是整數,所以沒變
print(float("12.5")) #雖然後面是字串,但前面加入了float,所以強制轉換,但原本就有小數點,所以沒變
print(eval("5.5*2")) #eval:計算輸入之值
```

#print的三種方式

#1.透過區隔要列印的變數eg:print("This,"is","a","book")
#2.利用字串的str.format()(新式)
#3.%格式化輸出字符串(舊式)

```python
print("This"+"is"+"a"+"book") #字串連著
print("This","is","a","book") #會區隔,為一般常見的
print("This") 
print("is")    #這四個都是單純印出每個顯示的字串樣子
print("a")
print("book")
```

#･利用字串的str.format(),在{}中替代變數的位置
#･eg:print("This{}a book".format("is"))
#･{}中可以設定格式
#･例如:{0:10}(10個空格,靠左對齊)
#･例如:{:,2f}(三個數字用,分隔,小數點後2位數)

```python
print("This {} a book".format("is")) #前面的{}裡要填的是呼應到後面format(格式)("is")
x=123456.7819
print("{:.2f}".format(x)) #小數點.後兩位
print("{:,.2f}".format(x)) #小數點前
```

```python
print("This %10s a book%-10s" % ("is","!")) #字串占10%,靠右2格
print("%7.3f" % (687.4333)) #字符寬度為7,保留3位小數
```
