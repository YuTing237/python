#函式
‧ 獨立出來的程式組

‧ 可以接收傳入值(多個)

‧ 輸出回傳值(一個)


```python
#傳統的寫法
import math
radius = eval(input("請輸入圓的半徑：")) #把輸入的字串變成數學計算型態
print("圓的面積為：",radius * radius * math.pi) #會印出圓的面積
```

#圓面積計算函式

```python
import math
def circle_area(redius): #radius為半徑
  area = radius * radius * math.pi #圓面積的公式:半徑 * 半徑 * pi
  return area

print("半徑10, 圓面積為:",circle_area(10))  #套入公式進行運算,會產生要的結果
print("半徑5, 圓面積為:",circle_area(5))
print("半徑20, 圓面積為:",circle_area(20))
```
