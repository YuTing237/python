#List[串列]

使用中括號[ ]包裹元素

```python
score=[90,100,88,60,79]
print(score[0])  #索引從0開始
print(score[1])  #索引從1開始

print(score[-1]) #倒數第一個
print(score[-2]) #倒數第二個 
```

```python
#score=[90,100,88,60,79]
for i in range(len(score)): #len(score): score的長度
  print(score[i])
```

```python
#score=[90,100,88,60,79]
print(score[0:3]) #取出score[0]-score[2],不包含score[3]
print(score[:]) #取出所有的值
print(score[:4]) #取出score[0]-score[3],不包含score[4]
print(score[2:]) #取出score[2]後面所有的值  
print(score[0:5:2]) #間隔2
print(score[::2])  #間隔2
print(score[::-2]) #間隔-2,代表從列表最後往前取值,間隔2
print(score[4:0:-2]) #從第4個開始取值,間隔2
```

#list的長度

```python
my_list0=[]   #空串列
print(my_list0)
my_list1=list() 
my_list2=['python','javascript','c#'] 
print(my_list2)  #會印出list2的所有資料
print(len(my_list2))  #len:計算list的長度
print(my_list2[2])   #會印出list2的第2個資料
```

#append附加, remove移除, insert插入, clear清除

```python
my_list2=['python','javascript','c#']
my_list2.append('java')  #append:附加在最後面 
print(my_list2)      #因為append的關西,所以會多加上java的資料在後面
my_list2.remove('java')   #remove:移除
print(my_list2)      #因為remove的關西,所以除了java的資料,list2的資料會印出來
my_list2.insert(0,'java') #insert:插入物件,需要指定索引位置
print(my_list2)      #因為insert的關西,指定了java要放在第0個位置
my_list2.clear()      #clear:清除所有物件
print(my_list2)       #因為clear的關西,所以所有資料會被刪除
```
