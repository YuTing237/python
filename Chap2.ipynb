#Chap2 流程控制

#if
#elif
#else

```python
a=3
b=4       
if a>b :     #運用if,else,輸入數字
  print("a>b")  #如果這邊對,後面的else就不會產生
else:
  print("b>a")  #如果前面不對,else就會產生
```

```python
a=3
b=3       #運用if,elif,else,輸入數字
if a>b :
  print("a>b") #如果這邊對,後面的elif,else就不會產生,然後因為是字串,所以不會產生數字,而是a>b
elif a==b:
  print("a=b") #如果前面不對,elif這邊符合,就會產生這條,後面的else就不會產生,然後因為是字串,所以不會產生數字,而是a=b
else:
  print("b>a") #如果前面都不對,就會產生這條,然後因為是字串,所以不會產生數字,而是a>b
```

```python
a=eval(input("請輸入數字a:")) #eval:計算輸入之值 
b=eval(input("請輸入數字b:")) #產生可以自己輸入數字
if a>b :      #運用if,elif,else
  print("a>b")
elif a==b:     #輸入完數字後,進行判斷,產生符合的結果
  print("a=b")
else:  
  print("b>a")
```

```python
grade=87 
print("成績評鑑:") 
if grade >= 90:
    print("優等")   #運用if,elif,else
elif grade >=80:
    print("甲等")
elif grade >=70:    #輸入的數字為多少,就會印出符合的結果
    print("乙等")
elif grade >=60:
    print("丙等")
else:
    print("不及格")
```

#for迴圈

```python
for i in range(10): #range(10):0,1,2....9
  print(i) #印出i
```

```python
for i in range(0,10,1): #range(0開始,9結束(不包含10),每次進1步)
  print(i) #印出i
```

#求1+2+3+...10的總和

```python
sum=0
for count in range(1,11,1): #從1開始，間隔1, 不包含11
  sum=sum+count

print(sum) #印出sum
```

```python
for c in "I am a boy": #給一個名稱,來產生I am a boy,會形成直立式的,但因為只有前後加入"",所以字母會單獨分開
  print(c) #印出c
```

```python
for score in [90,100,88,77]: #分數為90,88,77,會形成直立式
  print(score) #印出score
```

```python
for s in ["john","mary","tom","simpson"]:  #給一個名稱,來產生john,mary,tom,simpson,會形成直立式,因為有加入[],而且也是字串,所以字母不會單獨分開
  print(s) #印出s
```

#while迴圈

```python
i=1
sum=0
while i<11: #利用while進行運算
  sum=sum+i 
  i+=1

print(sum) #印出sum
```
