#檔案的開啟跟寫入
os套件有自己的開啟檔案函式open,但是MTA認證採用的是python內建的open

  ‧w = 可寫(預設建立檔案) 若檔案存在則會清除舊檔案 <br>
  ‧w+ = 可寫可讀(預設建立檔案) 若檔案存在則會清除舊檔案<br>
  ‧r = 可讀<br>
  ‧r+ = 可寫可讀 但寫入資料指向開頭 <br>
  ‧a = 可寫不能讀 寫入資料指向尾端 <br>
  ‧a+ = 可寫可讀 寫入資料指向尾端 <br>
  ‧b = 轉為2進位文件 <br>
  ‧ab+ = 2進位文件的讀寫 寫入資料指向尾端

```python
#a=open() #python內建檔案
#a=os.open() #os套件的開檔
```

```python
#新增檔案
a=open("a.txt","w")
a.write("This is a test1")
a.write("This is a test2\n") #\n為換行
a.write("This is a test3\n")
a.close() #在要離開執行區塊前,記得一定要呼叫close()把buffer裡命令寫入  #關閉檔案
```

#檢查檔案是否存在

```python
os.path.isfile("a.txt") #判斷a.txt這個檔案是否存在,如果存在就為true,不存在為false
```

#讀取檔案

```python
a=open("a.txt","r") #讀檔
print(a.readline()) #readline()一次讀取一行
print(a.readline())
a.close()
```

#加上檔案是否存在的判斷

```python
if os.path.isfile("a.txt"): #判斷a.txt檔案存在
  a=open("a.txt","r")    #會打開a的資料
  print(a.readline()) #readline()一次讀取一行
  print(a.readline()) 
  a.close()      #在要離開執行區塊前,記得一定要呼叫close()把buffer裡命令寫入
else:
  print("檔案不存在") #沒有就會產生"檔案不存在"
```

#split函式

```python
if os.path.isfile("a.txt"):  #判斷a.txt檔案是否存在
  a=open("a.txt","r")     #會打開a的資料
  for line in a:      #一行一行讀取a檔案
    print(line.split(" ")) #根據" "(空白)來切割每一行
  a.close()
else: 
  print("檔案不存在")    #沒有就會產生"檔案不存在"
```

#附加檔案

```python
a=open("a.txt","a") #在檔案的尾巴加上資料
a.write("This is a test4\n")
a.close()
```

```python
a=open("a.txt","r")
print(a.readlines()) #讀取所有的行
a.close()
```

#比較a與r+的不同
* a寫在檔案的結尾
* r+寫在檔案的開頭

```python
#如果需要附加在檔案最前方,要用r+參數
a=open("a.txt","r+") #在檔案的最前方加上資料
a.write("This is a test5\n")
a.close()
```

```python
a=open("a.txt","r")
print(a.readlines()) #讀取所有的行
a.close()
```

#刪除檔案

```python
if os.path.isfile("a.txt"):  #如果有a.txt的檔案
    os.remove("a.txt")   #就會進行刪除
    print("檔案刪除成功")  #刪除完會顯示"檔案刪除成功"
else:
    print("找不到檔案")   #如果沒有這個檔案就會顯示"找不到檔案"
print(os.path.isfile("a.txt"))
```
